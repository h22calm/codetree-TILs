```python
import cv2

image=cv2.imread('like_lenna.png',cv2.IMREAD_GRAYSCALE)
if image is not None:
  print("Image loaded successfully")
else:
  print("Failed to load image")
print(f"Variable type:{type(image)}")
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

```python
print(f"image shape:{image.shape}")
print(f"image size:{image.size}")
print(f"image data type:{image.dtype}")
```

```python
image_small=cv2.resize(image,(100,100))
cv2_imshow(image_small)
```

```python
print(f"image shape:{image_small.shape}")
print(f"image size:{image_small.size}")
print(f"image data type:{image_small.dtype}")
```

```python
image_quiz=cv2.resize(image,(250,250))
cv2_imshow(image_quiz)
```

```python
height,width=image_quiz.shape
matrix=cv2.getRotationMatrix2D((width/2,height/2),35,1)
result=cv2.warpAffine(image_quiz,matrix,(width,height),borderValue=0)
cv2_imshow(result)
```
